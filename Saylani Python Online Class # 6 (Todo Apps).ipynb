```python
# to-do application with elements in list(Simple)

elements = []           # elements is store in list

while True:             # we use true so loop will remain continue
    
    print("Elements in our list:")
    for element in elements:           # for printing of elements
        print(" -> " + element)
        
    print("\nOptions:")
    print("1. Add Element:")
    print("2. Delete Element:")
    print("3. Delete All Elements:")
    print("4. Exit:")
    print("Note: Please give inputs in numbers. eg: 1,2,3,4.")
    
    inp = int(input("Enter your choice: "))
    
    if inp == 1:
        elements.append(input("Enter your element: "))
        
    elif inp == 2:
        elements.remove(input("Enter elements to be deleted: "))
        
    elif inp == 3:
        elements.clear()
        
    elif inp == 4:
        break
        
    else:
        print("Invalid Input:")
        
    print("\n******************************************************************\n")    
    
    
```

```python
#to-do application with elements in dictionary

# Main menu
# Add tasks
# View tasks
# delete tasks
# terminate app

def IsInteger(userinput):     # to check user input is must be in integer or else run try condition
    try:
        int(userinput)
        return True
    except ValueError:    # except is a python keyword  same as catch in other language , ValueError is a keyword used in except
        return False

def Main_Menu(menu):
    print("\n==================\n   Main Menu:    \n==================\n    ")
    for menukey,menuname in menu.items():
        print("{}.{} ".format(menukey,menuname))
    usercommand = input("\nEnter Command:")
    return usercommand

def Add_Task(Task_Dictionary):
    print("\n==============================\n   Adding Task \n=================================\n")
    Task_Index = len(Task_Dictionary) + 1
    Task_Name = input("Task #{}:".format(Task_Index))
    Task_Dictionary.update({Task_Index : Task_Name})               # update is used for add task in dictionary
    print("Task #{} - Successfully Added!\n".format(Task_Index))
    
def View_Task(Task_Dictionary):
    print("\n==============================\n   Viewing Tasks \n=================================\n")
    if len(Task_Dictionary) > 0:
        for Task_Index,Task_Name in Task_Dictionary.items():
            print("Task #{} - {}".format(Task_Index,Task_Name))
    else:
        print("No Tasks Found!\n")

def Delete_Task(Task_Dictionary):
    print("\n==============================\n   deleting Tasks \n=================================\n")
    if len(Task_Dictionary) > 0:
        for Task_Index,Task_Name in Task_Dictionary.items():
            print("Task #{} - {}".format(Task_Index,Task_Name))
            
        Task_Index = int(input("Enter Task Index to be deleted: e.g:- 1 or 2 etc"))
        if Task_Index in Task_Dictionary.keys():
            Task_Dictionary.pop(Task_Index)
            print("Task #{} - Successfully Deleted!\n".format(Task_Index))
            
        else:
            print("Invalid Index! , Try Valid Index \n")
    
    else:
        print("No Tasks Found!\n")
        
        
menu = {
    1 : "Add Task",
    2 : "View Tasks",
    3 : "Delete Tasks",
    4 : "Terminate App"
}

Task_Dictionary={}

while(True):
    usercommand = Main_Menu(menu)
    
    if IsInteger(usercommand) == True:
        usercommand = int(usercommand)
        if usercommand > 0 and usercommand <= 4:
            print("Command Accepted")
            
            # Add Tasks
            if usercommand == 1:
                Add_Task(Task_Dictionary)
            
            # View Tasks
            if usercommand == 2:
                View_Task(Task_Dictionary)
            
            # Delete Tasks
            if usercommand == 3:
                Delete_Task(Task_Dictionary)
                
            # terminate app
            if usercommand == 4:
                print("\nApp Terminated:")
                break
                
        else:
            print("\nInvalid Command! , Try Valid Command")
    else:
        print("\nInvalid Command! , Try Valid Command")
```
